```python
###########Super Keyword

class Employee:
    
    raiseAmount=1.04
    numOfEmployee=0
    
    def __init__(self,first,last,pay):
        self.first=first
        self.last=last
        self.pay=pay
        self.email=first+'.'+last+'@company.com'
        Employee.numOfEmployee +=1
        
    def fullname(self):
        return '{} {}'.format(self.first,self.last)
    
    def apply_raise(self):
        self.pay=int(self.pay*self.raiseAmount)
        
        
class Developer(Employee):
    raiseAmount=1.10
    
    def  __init__(self,first,last,pay,prog_lang):
        super().__init__(first,last,pay) # let super class handel init mehtod
        self.prog_lang=prog_lang
        
        
class Manager(Employee):
    
    def  __init__(self,first,last,pay,employees=None):
        super().__init__(first,last,pay) # let super class handel init mehtod
        if employees is None:
            self.employees=[]
        else:
            self.employees=employees
            
    def add_emp(self,emp):
        if emp not in self.employees:
            self.employees.append(emp)
            
    def remove_emp(self,emp):
        if emp not in self.employees:
            self.employees.remove(emp)
            
    def print_emp(self):
        
        for emp in self.employees:
            print('-->',emp.fullname())
        
        
        
        
        
        
dev_1= Developer('Mrigank','Nandan',80000,'Python')
dev_2= Developer('Dewakar','Prasad',90000,'Scala')

mgr_1=Manager('Poonam','Prasad',100000,[dev_1])

print(mgr_1.email)


mgr_1.print_emp()
```

```python
###########SubClass

class Employee:
    
    raiseAmount=1.04
    numOfEmployee=0
    
    def __init__(self,first,last,pay):
        self.first=first
        self.last=last
        self.pay=pay
        self.email=first+'.'+last+'@company.com'
        Employee.numOfEmployee +=1
        
    def fullname(self):
        return '{} {}'.format(self.first,self.last)
    
    def apply_raise(self):
        self.pay=int(self.pay*self.raiseAmount)
        
        
class Developer(Employee):
    raiseAmount=1.10
    
    def  __init__(self,first,last,pay,prog_lang):
        super().__init__(first,last,pay) # let super class handel init mehtod
        self.prog_lang=prog_lang
        
        
class Manager(Employee):
    
    def  __init__(self,first,last,pay,employees=None):
        super().__init__(first,last,pay) # let super class handel init mehtod
        if employees is None:
            self.employees=[]
        else:
            self.employees=employees
            
    def add_emp(self,emp):
        if emp not in self.employees:
            self.employees.append(emp)
            
    def remove_emp(self,emp):
        if emp not in self.employees:
            self.employees.remove(emp)
            
    def print_emp(self):
        
        for emp in self.employees:
            print('-->',emp.fullname())
        
        
        
        
        
        
dev_1= Developer('Mrigank','Nandan',80000,'Python')
dev_2= Developer('Dewakar','Prasad',90000,'Scala')

mgr_1=Manager('Poonam','Prasad',100000,[dev_1])

print(isinstance(mgr_1,Manager))
print('**********')

print(isinstance(mgr_1,Employee))
print('**********')

print(isinstance(mgr_1,Developer))
print('**********')

print(issubclass(Developer,Employee))
print('**********')

print(issubclass(Manager,Employee))
print('**********')

print(isinstance(Manager,Developer))
print('**********')
```
